### Step 5. Feature Engineering

```python
# Print 5 rows of the dataset
df.head(5)
```

```python
# Display the column names of the DataFrame
df.columns
```

```python
# 1. id: Unique identifier for each house.
# 2. date: The date on which the house was sold.
# 3. price: The sale price of the house (in USD).
# 4. bedrooms : Number of bedrooms in the house.
# 5. bathrooms: Number of bathrooms in the house (can include half-baths).
# 6. sqft_living: Interior living space in square feet.
# 7. sqft_lot: Total lot size (land area) in square feet.
# 8. floors: Number of floors (levels) in the house.
# 9. waterfront: Indicates if the house has a waterfront (0 = no, 1 = yes).
# 10. view: Quality of the view from the house (0â€“4).
# 11. condition: Overall condition of the house (1â€“5).
# 12. grade: Overall grade of the house (based on King County grading system, 1â€“13).
# 13. sqft_above: Square footage of the house above ground level (no basement).
# 14. sqft_basement: Square footage of the basement.
# 15. yr_built: The year the house was originally built.
# 16. yr_renovated: The year the house was last renovated (0 if never renovated).
# 17. zipcode: The ZIP code where the house is located.
# 18. lat:  Latitude coordinate of the houseâ€™s location.
# 19. long:  Longitude coordinate of the houseâ€™s location.
# 20. sqft_living15: Average interior living space (in sqft) of homes in the same neighborhood (based on 15 nearest neighbors).
# 21. sqft_lot15: Average lot size (in sqft) of homes in the same neighborhood (based on 15 nearest neighbors).
# 22. year: Extracted sale year from the date column.
# 23. month: Extracted sale month from the date column (1â€“12).
# 24. day: Extracted sale day from the date column (1â€“31).
# 25. weekday: Extracted weekday from the date column (e.g., Monday=0, Sunday=6).
```

```python
# Create a new feature: house_age
df['house_age'] = df['year'] - df['yr_built']
```

```python
# Create a feature: price_per_sqft
df['price_per_sqft'] = df['price'] / df['sqft_living']
```

```python
# Seeing Information about the dataset
df.info()
```

```python
# Select numerical features (excluding the target 'price')
X = df.drop(columns=['price'])
y = df['price']
```

```python
# Compute mutual information scores
mutual_info = mutual_info_regression(X, y)
```

```python
# Convert to DataFrame for better visualization
feature_importance = pd.DataFrame({'Feature': X.columns, 'Importance': mutual_info})
feature_importance = feature_importance.sort_values(by='Importance', ascending=False)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importance['Importance'], y=feature_importance['Feature'], palette='viridis')
plt.xlabel('Mutual Information Score')
plt.ylabel('Feature')
plt.title('Feature Importance (Mutual Information)')
plt.show()
```

>Feature Importance (Mutual Information Score)

> Top influencing features on price: zipcode is the most informative feature, indicating location plays a crucial role in house prices. sqft_living, bathrooms, grade, and sqft_above also have high importance, meaning larger and better-graded houses tend to be more expensive. price_per_sqft has significant importance, which makes sense since price per area is directly linked to property value.

> Less influential features: Features like yr_renovated, waterfront, and sqft_basement show much lower mutual information scores, meaning they have minimal contribution to price prediction.

```python
# Select key numerical features for scatter plots
selected_features = ['sqft_living', 'grade', 'sqft_above', 'bathrooms', 'lat', 'long', 'sqft_living15']
```

```python
# Create scatter plots
plt.figure(figsize=(12, 8))
for i, feature in enumerate(selected_features, 1):
    plt.subplot(3, 3, i)
    sns.scatterplot(x=df[feature], y=df['price'], alpha=0.5)
    plt.xlabel(feature)
    plt.ylabel('Price')

plt.tight_layout()
plt.show()
```

> Weaker or unclear relationships: bathrooms vs. price shows some trend, but itâ€™s not as strong, suggesting more bathrooms slightly increase price but not linearly. lat & long vs. price reveal geographic clustering, meaning certain areas are more valuable, but latitude and longitude alone might not fully explain price variations.

### Step 6. Standard Scaling

```python
# Select numerical features that need scaling
features_to_scale = ['sqft_living', 'sqft_lot', 'sqft_above', 'sqft_basement',
                     'lat', 'long', 'sqft_living15', 'sqft_lot15', 'year', 'price_per_sqft']

# Initialize Standard Scaler
scaler = StandardScaler()

# Apply Standard Scaling to selected features
df[features_to_scale] = scaler.fit_transform(df[features_to_scale])

# Display the first 5 rows of the dataset
df.head()
```

```python
# Alternative: Scale all numerical columns except 'price'
features_to_scale = df.drop(columns=['price']).columns  # Drop 'price' before scaling

# Initialize Standard Scaler
scaler = StandardScaler()

# Apply Standard Scaling to all features except 'price'
df[features_to_scale] = scaler.fit_transform(df[features_to_scale])

# Display the first 5 rows of the dataset
df.head()
```

> I used two methods for scaling:
>
  > First Method: Only specific features are scaled, while others remain in their original form.
  >
  > Second Method: All numerical columns except price are scaled to ensure consistency across features.


# Step 7. Model training

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python

d_tree = DecisionTreeRegressor(random_state=42)
d_tree.fit(X_train, y_train)
y_pred = d_tree.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse:.2f}')
print(f'R-squared: {r2:.2f}')
     
```

```python

# Perform K-fold cross-validation (k=5)

cv_scores = cross_val_score(d_tree, X, y, cv=5)  # cv=5 means 5-fold cross-validation
avg_cv_scores=np.mean(cv_scores)

avg_cv_scores
```
